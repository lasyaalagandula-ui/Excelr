```python
#1.Load the dataset into a data analysis tool or programming environment (e.g., Python with pandas library).
import pandas as pd
df=pd.read_csv("sales_data_with_discounts.csv")
df
```

```python
#Identify numerical columns in the dataset.
df1=df.describe(include="number")
df1
```

```python
#2.Calculate the mean, median, mode, and standard deviation for these columns.
for col in df1.columns:
    print(f"mean values of {col} are:")
    print(df1[col].mean())
```

```python
df.select_dtypes(include="number") #we use delect_types for seperate datatypes
```

```python
for col in df1.columns:
    print(f"median for the {col} are:")
    print(df1[col].median())
```

```python
for col in df1.columns:
    print(f"Standard deviation for {col} are:")
    print(df1[col].std())
```

```python
for col in df1.columns:
    print(f"mode for the {col} are:")
    print(df1[col].mode())
```

# 3.Provide a brief interpretation of these statistics.
###### The mean shows the average value of each column. It tells us the overall average sales, price, and discount in the dataset.
###### The median shows the middle value of the data. It helps us understand the central value and reduces the effect of very high or very low values.
###### The standard deviation shows how much the values vary from the average. A high standard deviation means the values change a lot, while a low standard deviation means the values are more consistent.
###### The minimum and maximum values show the lowest and highest values in the dataset. A big difference between them means there is a large variation in the data.
###### The mode shows the most frequently occurring value in a column.
###### Therefore, the statistics help us understand the distribution, variation, and central tendency of the sales and discount data.


```python
#Data Visualization
import matplotlib.pyplot as  plt #we use this to show graphs seperately
for col in df1.columns:
    print(f"the histograph of {col} is:")
    df1[col].plot(kind='hist')
    plt.show()
#As the skewness is right side or positive therefore,most values are low to moderate and less values are high
#the presence of a few extremely high values suggests there may be outliers in the dataset
#Overall,the data is not evenly distributed. Most transactions have moderate values, but a few large values increase the spread of the data.    
```
