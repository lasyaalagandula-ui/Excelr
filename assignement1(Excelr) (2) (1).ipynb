```python
#1.Write a Python program that checks whether a given number is prime or not.
n=int(input())
if n<=1:
    print("Not a prime number")
for i in range(2,n):
    if n%i==0:
       print("Not a prime number")
       break
else:
    print("It is a prime number")
        
    
    
```

```python
#3 Squares of Even/Odd Numbers
print("Even Numbers")
for i in range(100,201):
    if i%2==0:
        print(i,i**2)
print("odd Numbera")
for i in range(100,201):
    if i%2!=0:
        print(i,i**2)
```
